# Internet x Moradia

```python
sns.countplot(x="internet", hue="address", data=merge)
merge['internet'].value_counts()
```

```python
rural = merge['address'].value_counts()['R']
urbano = merge['address'].value_counts()['U']
semrural = merge[(merge.address == "U") & (merge.internet == 0)].count()
semurbano = merge['internet'].value_counts()[0] - semrural

print(semrural / rural)
print(semurbano / urbano)
```

Ainda que a amostra de estudantes vivendo no espao rural seja pequena (79), podemos observar que a proporo de estudantes sem acesso  internet  muito maior no espao rural (39,2%) do que no espao urbano (0,08%). 

# Tamanho da familia x Moradia
